#Text preprocessing 

```python
with open('/content/drive/MyDrive/Colab Notebooks/contractions.json', 'r') as f:
    contractions_dict = json.load(f)
contractions = contractions_dict
stop = stopwords.words('english')
lemmatizer = WordNetLemmatizer()
spell = SpellChecker(language='en')
new_stop=[]
not_removable_stop=['but','not','few','more','most','no','nor','very','too','same','only']
for e in stop:
        if e not in not_removable_stop:
            new_stop.append(e)
            
def cleaned_text(text):                                            
    text = re.sub(r"\d+", " ", str(text))                                                                       
    text = re.sub(r"\b[a-zA-Z]\b", "", str(text))                  
    text = re.sub(r"[^\w\s]", " ", str(text))                                            
    text = re.sub(r"\s+", " ", str(text))
    text = text.lower()
    for word in text.split():
        if word.lower() in contractions:
            text = text.replace(word, contractions[word.lower()]) 
    words = word_tokenize(text)
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    text = ' '.join(lemmatized_words)
    text = ' '.join([word for word in text.split() if word not in (new_stop)])
    for i, word in enumerate(text.split()):
        corrected_word = spell.correction(word)
        if corrected_word and corrected_word != word:
            text = text.replace(word, corrected_word)
    return text
df2['product_review'] = df2['product_review'].apply(cleaned_text)
#df2.sample(5)
```

#RNN code for sentiment analysis using LSTM

```python
from keras.preprocessing.text import Tokenizer
from keras_preprocessing.sequence import pad_sequences
from keras.layers import Dense, Embedding, LSTM, SpatialDropout1D
from keras.models import Sequential
from sklearn.model_selection import train_test_split
from keras.utils.np_utils import to_categorical
from imblearn.under_sampling import RandomUnderSampler
import keras
import tensorflow
from keras.optimizers import Adam
review = df2['product_review'].to_list()

# Tokenization and Padding
max_fatures = 71000
tokenizer = Tokenizer(num_words=max_fatures, split=' ')
tokenizer.fit_on_texts(review)
X = tokenizer.texts_to_sequences(review)
X = pad_sequences(X)

# Train-Test Split
y = df2['product_rating']
print(X.shape,y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

y_train = keras.utils.to_categorical(y_train-1, num_classes=5)
y_test = keras.utils.to_categorical(y_test-1, num_classes=5)

model = Sequential()
model.add(Embedding(max_fatures, 128, input_length=X_train.shape[1]))
model.add(LSTM(64, dropout=0.7, recurrent_dropout=0.5))
#model.add(LSTM(32, dropout=0.8, recurrent_dropout=0.8))
model.add(Dense(5, activation='softmax'))

# Model Training
batch_size = 220
epochs = 20
model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=0.002), metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size,validation_data=(X_test, y_test),validation_split=0.2)

# Model Evaluation
score, acc = model.evaluate(X_test, y_test, batch_size=batch_size)
print("Score: %.2f" % (score))
print("Accuracy: %.2f" % (acc))
```

#Random Forest for sentiment analysis 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from imblearn.over_sampling import RandomOverSampler

# Divide the data into training and testing set
X_train, X_test, y_train, y_test = train_test_split(df2['product_review'], df2['product_rating'], test_size=0.3, random_state=42)

# Convert the text into numerical data using TfidfVectorizer with n-grams
vectorizer = TfidfVectorizer(ngram_range=(1,2))
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)

# Train the Random Forest model
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)

# Predict the sentiment on test data
rf_predictions = rf_model.predict(X_test)

# Calculate the accuracy of the Random Forest model
print("Accuracy (Random Forest): ", accuracy_score(y_test, rf_predictions))

print(precision_recall_fscore_support(y_test, rf_predictions, average='macro'))
print(metrics.classification_report(y_test, rf_predictions))

```

```python
from sklearn.model_selection import GridSearchCV

# Define a range of parameters to search over
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Initialize a grid search object with the classifier and parameter grid
grid_search = GridSearchCV(rf_model, param_grid, scoring='accuracy', n_jobs=-1)

# Fit the grid search to the training data
grid_search.fit(X_train, y_train)

# Get the best parameters from the grid search
best_params = grid_search.best_params_

# Train a new random forest classifier with the best parameters
rf_model_tuned = RandomForestClassifier(**best_params)
rf_model_tuned.fit(X_train, y_train)

# Predict the sentiment on the test data with the tuned model
rf_predictions_tuned = rf_model_tuned.predict(X_test)

# Calculate the accuracy of the tuned model
print("Accuracy (Tuned Random Forest): ", accuracy_score(y_test, rf_predictions_tuned))
```
